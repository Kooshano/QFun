# QFun-KAN Signed Modes Demo

Focused walkthrough of Mode A and Mode B signed encoding APIs.

```python
import numpy as np
import matplotlib.pyplot as plt

from qfun_layers.amplitude_encoding import create_grid
from qfun_layers.signed_encoding import (
    mode_a_signed_encode,
    reconstruct_mode_a_signed,
    mode_b_signed_decompose,
    reconstruct_mode_b_signed,
)
```

```python
x = create_grid(-1, 1, n_qubits=5)
f = np.sin(2*np.pi*x) + 0.2*x
plt.figure(figsize=(6,3))
plt.plot(x, f)
plt.title('Mixed-sign target')
plt.show()
```

```python
alpha, sign_bits = mode_a_signed_encode(f)
p_mag = np.abs(alpha)**2
p_plus = np.where(sign_bits==0, p_mag, 0.0)
p_minus = np.where(sign_bits==1, p_mag, 0.0)
q_hat_a = reconstruct_mode_a_signed(alpha, sign_bits)

plt.figure(figsize=(7,3))
plt.bar(x, p_plus, width=(x[1]-x[0])*0.8, alpha=0.6, label='p_plus')
plt.bar(x, -p_minus, width=(x[1]-x[0])*0.8, alpha=0.6, label='-p_minus')
plt.plot(x, q_hat_a, 'k--', lw=2, label='q_hat')
plt.legend()
plt.title('Mode A: ancilla-separated histogram')
plt.show()
```

```python
p_plus_b, p_minus_b, z_plus, z_minus = mode_b_signed_decompose(f)
q_hat_b = reconstruct_mode_b_signed(p_plus_b, p_minus_b, z_plus, z_minus)

print('Mode B reconstruction max abs error:', np.max(np.abs(q_hat_b - f)))

plt.figure(figsize=(6,3))
plt.plot(x, f, label='original')
plt.plot(x, q_hat_b, '--', label='reconstructed')
plt.legend()
plt.title('Mode B: q_hat = Z+ p+ - Z- p-')
plt.show()
```
